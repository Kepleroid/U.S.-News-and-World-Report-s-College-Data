# <font  color='red'>Principal Component Analysis</font>

```python
def dot(x, y):
    return sum(x[i] * y[i] for i in range(len(x)))
```

```python
# Standard Deviation

def mean(x):
    return sum(x) / len(x)

def de_mean(x):
    """translate x by subtracting its mean (so the result has a mean 0)"""
    x_bar = mean(x)
    return [x_i - x_bar for x_i in x]

def sum_of_squares(x):
    return sum(pow(x_i, 2) for x_i in x)

def variance(x):
    """assumes x has at least two elements"""
    n = len(x)
    deviations = de_mean(x)
    return sum_of_squares(deviations) / (n-1)

def standard_deviation(x):
    return math.sqrt(variance(x))
    
```

```python

def standardise_data(data_matrix, num_cols):

    """returns the menas and standard deviations of each column"""
    means = [mean(get_column(data_matrix, j))
            for j in range(num_cols)]
    stdevs = [standard_deviation(get_column(data_matrix, j))
             for j in range(num_cols)]
    for row in range((data_matrix.shape)[0]):
        for col in range((data_matrix.shape)[1]):
            data_matrix[row][col] = (data_matrix[row][col] - means[col]) 
            data_matrix[row][col] = data_matrix[row][col] / stdevs[col]
    return data_matrix

def get_column(matrix, col):
    return [matrix[i][col] for i in range((matrix.shape)[0])]


def make_matrix(num_rows, num_cols, entry_fn):
    """returns a num_rows x num_cols matrix
    whose (i, j)th entry is entry_fn(i, j)"""
    return [[entry(i, j)
            for j in range(num_cols)]
           for i in range(num_rows)]

x = np.array(x)
standardise_data(x, (x.shape)[1])

    
```

# 1) Eigendecomposition - Computing Eigenvectors and Eigenvalues



```python
def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n - 1)

def covariance_matrix(matrix, num_rows, num_cols):
    mean_vec = np.mean(matrix, axis=0)
    cov_mat = (x - mean_vec).T.dot((x - mean_vec)) / (x.shape[0]-1)
    return cov_mat
#     for col in range(num_cols):
#         for row in range(num_rows):
#             matrix[row][col] = covariance(matrix[row], matrix[col])
#     return matrix
            
cov_mat = covariance_matrix(x, 17, 17)
print(cov_mat)
print(cov_mat.shape)
    
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('Eigen Values: \n', eig_vals)
print('Eigenvectors: \n', eig_vecs)
```

# Selecting Principal Components

```python
for ev in eig_vecs:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Everything ok!')
```

In order to decide which eigenvecto can be dropped without losing too much information fr the construction of lower-dimension subspace, we need to inspect the corresponding eigenvalues: the eigenvectors with the lwest eigenvalues bea the least information about the distribution of the data; those are the ones that can be dropped

In order to do so, the common approach is to rank the eigenvalues from highest to lowest in order to choose the top k eigenvectors


```python
#make a list of (eigenvalue, eignvector) tuples
eig_pairs = [(np.abs(eig_vals[i]),
             eig_vecs[i]) for i in range(len(eig_vals))]

eig_pairs.sort()
eig_pairs.reverse()
print('Eigenvalues in descending order: ')
print(eig_pairs)
for i in eig_pairs:
    print(i[0])
    
```

After sorting the eignpairrs, the next question is "how many principal components are we going to choose for our new feature subspace?" A useful measure is the so-called "explained variance," which can be calculated from the eigenvalues. the eplained variance tells us how much information (variance) can be attributed to each of the principal components.

```python
total = sum(eig_vals)
var_exp = [(i / total)*100 for i in sorted(eig_vals, reverse=True)]
cum_ver_exp = np.cumsum(var_exp)

fig, ax = plt.subplots(figsize=(16,8))
ax.bar(list(range(len(var_exp))), var_exp)
ax.set_xlabel('Principal Components')
ax.set_ylabel('Variance')
ax.set_xlim([0, 16])
labels = [item.get_text() for item in ax.get_xticklabels()]
for i in range(len(labels)):
    labels[i] = "PC " + str(i*2)

ax.set_xticklabels(labels)

plt.plot(list(range(len(var_exp))), var_exp, color='r', linewidth=5, label='Cumulative explained variance')
plt.legend()
plt.show()

```

# <font color='red'> K - Means Clustering </font>

```python
import random
from statistics import mean

class K_Means():
    def __init__(self, k=2, tolerance=0.001, hm_epochs=300):
        self.k = k
        self.tol = tolerance
        self.hm_epochs = hm_epochs
        
    def classify(self, predict):
        distances = [np.linalg.norm(centroid - predict) for centroid in self.centroids]
        classification = distances.index(min(distances))
        return classification
    
    def fit(self, input_data):
        self.centroids = {}
        
        for i in range(self.k):
            self.centroids[i] = input_data[i]
        
        for epoch in range(self.hm_epochs):    
            self.classifications = {}
            
            for i in range(self.k):
                self.classifications[i] = []
            
            for featureset in input_data:
                distances = [np.linalg.norm(featureset - self.centroids[centroid]) for centroid in self.centroids]
                index_value = distances.index(min(distances))
                self.classifications[index_value].append(featureset)
                
            prev_centroids = dict(self.centroids)
            
            for classification in self.classifications:
                self.centroids[classification] = np.average(self.classifications[classification], axis=0)
                
            diffs = [np.sum((self.centroids[i] - prev_centroids[i])/prev_centroids[i]*100.0) for i in self.centroids]
            if max(diffs) < self.tol:
                break
                
#             optimized = True
            
#             for c in self.centroids:
#                 original_centroid = prev_centroids[c]
#                 new_centroid = self.centroids[c]
#                 if np.sum((new_centroid-original_centroid)/original_centroid*100.0) > self.tol:
#                     print(np.sum((new_centroid-original_centroid)/original_centroid*100.0))
#                     optimized = False
                    
#             if optimized:
#                 print(self.centroids)
#                 break

                
                
X = np.array([[1, 2],
              [1.5, 1.8],
              [5, 8 ],
              [8, 8],
              [1, 0.6],
              [9,11]])  
colors = 10*["g","r","c","b","k"]

                
clf = K_Means()
clf.fit(X)

fig, ax = plt.subplots(figsize=(10, 10))

for centroid in clf.centroids:
    ax.scatter(clf.centroids[centroid][0], clf.centroids[centroid][1],
                marker="o", color="k", s=150, linewidths=5)

for classification in clf.classifications:
    color = colors[classification]
    for featureset in clf.classifications[classification]:
        ax.scatter(featureset[0], featureset[1], marker="x", color=color, s=150, linewidths=5)
        
plt.show()
    
    
    
    
```

# Accuracy

```python
def test_accuracy(centroids, x_test, y_test):
    classify = {}
    for i in range(len(y_test)): #change the actual data into numerical data
        if y_test[i] in classify:
            y_test[i] = classify[y_test[i]]
        else:
            classify[y_test[i]] = i
            y_test[i] = classify[y_test[i]]
    y_pred = []
    for i in range((x_test.shape)[0]):#loop through each instance in data
        distances = [np.linalg.norm(x_test[i] - centroid) for centroid in centroids]
        index_value = min(distances)
        classification = distances.index(index_value)
        y_pred.append(classification)
        
    correct = 0
    for i in range(len(y_pred)):
        if y_pred[i] == y_test[i]:
            correct += 1
    return (correct / len(y_pred)) *100

        
        
test_accuracy(clf.centroids, x, y_list)

    
```
